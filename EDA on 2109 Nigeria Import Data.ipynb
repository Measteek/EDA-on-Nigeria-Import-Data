```python
data.drop(['S/N', 'SCANNING_COY', 'PREFIX', 'BANK_CODE', 'BANK_NUMBERING', 'MFNUM', 'APP_YEAR', 'EXC_RATE', 'FREIGHT_VAL', 'CFR', 'USD EQUIV', 'APPLICANT_RCN_CODE', 'IMPORTER_ADDRESS', 'IMPORTER_EMAIL', 'EXPORTER_ADDRESS', 'EXPORTER_EMAIL', 'Bank Code'], axis=1, inplace=True)
```

```python
data.head()
```

```python
#total sales quantities by country
data.groupby('APPLICANT_NAME').ITEM.sum().sort_values(ascending=False)[:10]
```

```python
top_ten = data.groupby('ITEM').FOB_Value_Naira.sum().sort_values(ascending=False)[:10]
top_ten
```

```python
#use horizontal bar chart to visualize top 10
#use the series index as vertical axis and series values as horizontal axis of the barh() plot
plt.figure(figsize=(15,10))
plt.barh(y=top_ten.index, width=top_ten)
plt.show()
```
